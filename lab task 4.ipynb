```python
n=int(input("Enter the size of the chessboard: "))
pos=[]
print("enter the column number for the queen in each row(start from 0):")
for i in range(n):
    c=int(input( "column for row" + str(i)+": "))
    pos.append(c)
    Valid=True  # check
for i in range(n):
    for j in range(i+1,n):
        if pos[i]==pos[j] or  abs(pos[i]-pos[j])==abs(i-j):
            Valid=False
        
if Valid:
    print("\n True")
else:
    print("\n False")
horizontal_line="+"+"---+"*n
print("\nYour chessboard is:\n")  
for i in range(n):  #rows
    print(horizontal_line)
    for j in range(n):   #columns
        if j == pos[i]:
            print("Q",end=" | ")
        else:
            print(" ",end=" | ")
    print()  
```
