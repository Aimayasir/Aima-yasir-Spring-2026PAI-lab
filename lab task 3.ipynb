```python
jA=4
jB=3
visited = set()

def dfs(a , b):
    if (a, b) in visited:
        return False
    visited.add((a, b))

    print(f"current state: jA : {a}, jB : {b}")
    if a == 2 or b == 2:
        print("Goal reached")
        return True
    
# Now apply all possible actions
    if dfs(jA,b):
        print("Fill first jug")
        return True
    if dfs(a,jB):
        print("Fill second jug")
        return True
    if dfs(0,b):
        print("Empty first jug")
        return True 
    if dfs(a,0):
        print("Empty second jug")
        return True
    
    pour = min(a, jB - b)
    newA = a - pour
    newB = b + pour
    if dfs(newA, newB):
        print("Pour from first jug to second jug")
        return True

    pour = min(b, jA - a)
    newA = a + pour
    newB = b - pour
    if dfs(newA, newB):
        print("Pour from second jug to first jug")
        return True
     
    return False
dfs(0,0)    
```
